## Importing neccessary libraries

```python
import pandas as pd #For Data manupulation and Analysis
import numpy as np #For numeric calculation
import matplotlib.pyplot as plt #For Data visualization
import seaborn as sns #For Data visualization
%matplotlib inline
```

## Reading Dataset

```python
df = pd.read_csv('Breast_Cancer.csv')
df.head()
```

```python
#Dropping column 'id' and 'Unnamed: 32' from Dataset

df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
#Information os DataFrame
df.info()
```

```python
#Numerical distribution of Data
df.describe()
```

## Visualize the Data

```python
sns.pairplot(df, hue='diagnosis')
```

```python
#Count the Target class

sns.countplot(df['diagnosis'])
```

```python
#Converting the target variable into numerical category

df['diagnosis'] = pd.get_dummies(df['diagnosis'], drop_first=True, prefix=None)
```

```python
df.head()
```

```python
#Heatmap of Correlation matrix
plt.figure(figsize=(22,22))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=2)
```

```python
X= df.drop('diagnosis', axis=1)
y= df['diagnosis']
```

```python
#Viasualise correlation of Input features with Target feature
plt.figure(figsize=(15,5))

ax = sns.barplot(X.corrwith(y).index, X.corrwith(y))
ax.tick_params(labelrotation= 90)
```

```python
X.head()
```

```python
#Split DataFrame into Train and Test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Model building

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

## Support Vector Classifier

```python
from sklearn.svm import SVC
svc_classifier= SVC(kernel='rbf')
```

```python
svc_classifier.fit(X_train, y_train)
```

```python
y_pred= svc_classifier.predict(X_test)
```

```python
cm= confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_classifier= LogisticRegression()
```

```python
lr_classifier.fit(X_train, y_train)
```

```python
y_pred= lr_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```

## K-Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Choosing K-value

err_rate= []
for i in range(1,41):
    knn_classifier= KNeighborsClassifier(n_neighbors=i)
    knn_classifier.fit(X_train, y_train)
    y_pred= knn_classifier.predict(X_test)
    
    err_rate.append(np.mean(y_test != y_pred))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,41), err_rate, marker='o', linestyle='dashed', markersize=8, color='red')
plt.xlabel('K-values')
plt.ylabel('Error rate')
```

```python
#Taking K value as 5

knn_classifier= KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
y_pred= knn_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```
